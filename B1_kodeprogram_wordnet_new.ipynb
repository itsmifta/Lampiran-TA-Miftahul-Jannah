```python
import pandas as pd
import textdistance
import datetime
import os
from itertools import islice
```

```python
# filtering word which in wordnet synset or not

def filtering(in_f, out_f, size):
    from nltk.corpus import wordnet as wn
    print(datetime.datetime.now())
    reader = pd.read_csv(in_f, delimiter=',', sep="\n", chunksize=size)
    ch =0
    for chunk in reader:
        df_filtering = pd.DataFrame({'code': [],'ingredient': [],}) 
        for index, row in chunk.iterrows():
            if type(row['ingredient']) is str: # memfilter apakah word berupa string
                if wn.synsets(row['ingredient'], 'n'): # memfilter apakah word ada di wordnet (noun)
                    df_filtering = df_filtering.append({'code': row['code'],'ingredient': row['ingredient'],}, ignore_index=True)
        df_filtering.to_csv(out_f, index=False, header=False, mode='a')
        print(datetime.datetime.now())
        ch+=1
        print(ch)
```

```python
filtering('tokenized_nodup_tahap2.csv', 'tokenized_filtered.csv',1000)
```
